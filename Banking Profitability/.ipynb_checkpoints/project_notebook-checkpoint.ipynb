# Research Question

What are the key determinants in predicting bank profitability?

# Motivation

The banking sector is an important component of every economy and has a great impact on national financial stability and economic growth. Studies show that profitability is significantly related to stability, as banks are able to build financial buffers from their net income (Beck & Casu, 2016). Currently, there are many innovative players that provide unbundled banking services, thus creating a highly competitive environment. Contestability theory predicts that more competitive markets result in lower profits and related financial stability associated with the banking sector. It is important to understand how banking performance has changed and what are its main drivers. Policy makers need to be able to make accurate decisions to influence the banking sector (Bahar & Tahir, 2009), providing the main motivation of our research.






Beck, T. & Casu, B. (Eds.). (2016). The Palgrave Handbook of European Banking. Palgrave Macmillan.

Bakar, N. M. A., & Tahir, I. M. (2009). Applying multiple linear regression and neural network to predict bank performance. International Business Research, 2(4), 176-183.

# Model

The aim of the research is to identify drivers of bank profitability using US banking data. We follow the methodology from BIS (2018) and Bahar and Tahir (2009) by selecting determinants based on data availability. The following model is identified for the banking sector in country *i*:

$$ ROA_i =  β_0 + β_1LG_i + β_2GDP_i + β_3SIZE_i + β_4SR_i + β_5LIQ_i + β_6CR_i + \epsilon_i   $$

where

ROA = Return on average assets as a measure of profitability

LG = Loan asset base growth rate

GDP = GDP growth rate

SIZE = Size of the banking sector proxied by total assets

SR = Short-term interest rate proxied by a 3-month interbank rate

LIQ = Liquidity position of a bank captured by ratio of securities holdings (liquid asset) to total assets

CR = Loan loss provisions ratio to total loans as a measure for credit risk 



BIS (2018). Determinants of bank profitability in emerging markets. Kohlscheen, E., Murcia Pabón, A., & Contreras, J. BIS Working Papers.

# Data Description

The data used in the analysis is from the Federal Reserve Economic Data (FRED) database which is accessed by using Pandas Data Reader API. The FRED database provides information for all U.S. banks, thus it is a great representation of the US banking sector. Furthermore, the U.S. economy and its banking sector makes up a large share of the global banking environment. However, it does not account for individual bank characteristics since it is time-series data but not cross-sectional. Selected data variables are based on the Model explained above. We create Liquidity and Credit Risk ratios by dividing securities by total assets and provision of loans by loans, respectively. More details about each indicator and how it is denominated are listed below.

Data is standardized in the following ways. Most indicators are measured on a quarterly basis, however Loans and Securities have a monthly frequency, thus we convert it into a quarterly figure by taking every third value. We convert absolute variables to billions of dollars and percentage rates (including outcome variable Return on Assets) are converted decimal units. GDP is imported as stock variable and is transformed into GDP growth rate.  

Data is stored in a Pandas Data Frame tabular data format and columns are renamed to have more meaningful names. To get a quick overview of the dataset, we display the head or tail rows as well as the shape of the data table. The data covers a wide range from 1984 to 2020 - providing enough variation for our analysis.

When inspecting the data we primarily employ visual tools to get an overview, quickly identify patterns, and potential relationships. Interactive plots over the analyzed period are created to allow the user to manipulate the graphs. For further analysis we use neural network, random forest, and Bayesian linear regression for estimating most important banking profitability determinants.

```python
# Import variables from FRED from 1984 to 2020
start = datetime.datetime(1984, 1, 1)
end = datetime.datetime(2020, 7, 1)

df_q = pdr.DataReader(['USROA', 'LLBACBQ158SBOG', 'QBPBSTAS', 
                     'IR3TIB01USQ156N', 'GDP', 
                      'QBPQYLNLOSS'], 'fred', start, end)

df_m = pdr.DataReader(['INVEST', 'LOANS'], 'fred', start, end)
```

- The securities and loans data are imported seperately since it is monthly and not quarterly like the other vairables. 

```python
# Reset index for quarterly data
df_q.reset_index(inplace = True)
df_q.head()
```

```python
# Convert monthly data to quarterly by taking every 3rd observation, reset index
df_m = df_m.iloc[::3, :]
df_m.reset_index(inplace = True)
df_m.tail()
```

- Every third observation from the monthly data on securities and loans is kept to reduce the frequency of the data to quarterly.

```python
# Merge converted monthly data with quarterly data
df_i = df_q.merge(df_m, how = 'left', on = ['DATE'])
df_i.tail()
```

```python
# Rename columns 
df_i = df_i.rename(columns={'DATE':'Date', 'USROA':'ROA', 'LLBACBQ158SBOG':'Loan Growth Rate', 
                        'QBPBSTAS':'Total Assets', 'IR3TIB01USQ156N':'Interbank Rate', 
                        'QBPQYLNLOSS':'Provision Loans', 'INVEST':'Securities', 
                        'LOANS':'Loans'})
```

```python
# Convert date column to datetime format
df_i.Date = pd.to_datetime(df_i.Date)
```

```python
# Convert total assets and provision of loans to billions of dollars from millions
df_i['Total Assets'] = df_i['Total Assets']*0.001
df_i['Provision Loans'] = df_i['Provision Loans']*0.001
```

- Total assets and provision of loans are converted to billions of dollars from millions to match the units of securities and loans. 

```python
# Convert ROA, loan growth rate, and interbank rate from percent to decimal
df_i['ROA'] = df_i['ROA']*0.01
df_i['Loan Growth Rate'] = df_i['Loan Growth Rate']*0.01
df_i['Interbank Rate'] = df_i['Interbank Rate']*0.01
```

- ROA, loan growth rate, and interbank rate are converted from percentages to decimal units to match the units of liquidity and credit risk created below. 

```python
# Create GDP growth rate 
df_i['GDP Growth Rate'] = [(df_i.GDP[0] - 3794.706)/3794.706] + [(df_i.GDP[i + 1] - df_i.GDP[i])/df_i.GDP[i] for i in range(len(df_i.GDP) - 1)]
```

- GDP growth rate is calculated by the change from year to year. The value for the last quarter in 1983 (3794.706) is not imported but is available from the FRED. Without this value the conversion from GDP to GDP growth rate would be missing a value for the first quarter of 1984. 

```python
# Create variables liquidity and credit risk
df_i['Liquidity'] = df_i['Securities']/df_i['Total Assets']
df_i['Credit Risk'] = df_i['Provision Loans']/df_i['Loans']
```

- A proxy for liquidity is generated by dividing securities by total assets.
- A proxy for credit is calculated by dividing provision of loans by loans. 

```python
# Keep only variables used in the model
df = df_i[['Date', 'ROA', 'Loan Growth Rate', 'GDP Growth Rate', 'Total Assets', 
           'Interbank Rate', 'Liquidity', 'Credit Risk']]
# df.set_index('Date', inplace = True)
df.head()
```

# Method Explanation

The first method is regression analysis using a neural network to analyze the relationship between bank profitability (return on assets) and the 6 predictors. This method is based on correlations. It functions by having 6 input nodes connected to a hidden layer with multiple nodes and single output node. The models are trained with the features data as inputs and the single target variable as output. For each section both a full model and incomplete model, without total assets, are trained and fitted with the data. The data is normalized instead of standardized since normalization give better results when data is not normally distributed. For the first models a simple split of 80 percent for the training set and the other 20 percent for the validation set is used. Then a time series split method is used to train the models which uses a rolling window over several splits to specify the training and validation sets. The fit of the models is determined by calculating the root-mean-square error of their predictions for ROA. After, for comparison to the neural network regression models, random forest regression is used for both a full model and an incomplete model. It's straight forward to check feature importance from a random forest regression model to compare to the less certain sum of weights from input nodes in the neural network models. 

The second method to capture the effect of determinants on the outcome variable of profitability is Bayesian linear regression. The main advantage of the Bayesian processing, is that one is able to recover the distribution of each parameter rather than just a point estimate. All predictors are assumed to have a linear relationship with profitability and their combination (together with the constant and noise term) explains the variation in the outcome variable. This method is also based on correlations. Variables are standardized to help the algorithm. We provide conservative priors in the model (zero expectation) to avoid potential overfitting. In particular, model parameters are assumed to be normally distributed with mean 0 and standard deviation equal to 0.1. Trace is sampled from 2 chains for 1000 tunes and 10 000 iterations. Monte Carlo Markov chain (MCMC) algorithm is used to fit the model. Posterior predictive distribution is investigated to check if the algorithm has successfully converged and both chains report similar results. Next an alternative reduced model is analyzed to see if it provides higher predictive information, as part of the robustness check. The two models are compared based on Widely Applicable Information Criterion (WAIC). Results of posterior distribution of each determinant are analyzed from the chosen model. Finally, for sensitivity analysis, we generate random data ourselves and inspect how close the estimated slopes are from their true known values.

The three regressions methods are used as a robustness check and Bayesian regression is used over Lasso regression since there are several parameters in the model and Lasso regression performs better with fewer parameters.

- This section runs the full model as defined above with a train and validation split of 80:20 without shuffling. 

```python
# Subset variables from dataframe into features and target arrays
features = np.array(df[['Loan Growth Rate', 'GDP Growth Rate', 'Total Assets', 
                        'Interbank Rate', 'Liquidity', 'Credit Risk']])
target = np.array(df['ROA'])
features.shape, target.shape
```

- Features are defined as the 6 predictor variables
- Target is ROA

```python
# Normalize data 
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
features_scaled = scaler_x.fit_transform(features)
target_scaled = scaler_y.fit_transform(target.reshape(-1, 1))
```

- The features and target are normalized separately to increase the performance of the model

```python
# Split data without shuffling (80:20)
x_train, x_val, y_train, y_val = train_test_split(features_scaled, 
                                                  target_scaled, 
                                                  test_size=0.2, 
                                                  shuffle = False)
```

- The data is split into train and validation sets 'x' referring to the features which are inputs and 'y' referring to the output

```python
# Define function for keras sequential model architecture
def model(input, hidden, learning_rate):
    model = Sequential()
    model.add(InputLayer(input))
    model.add(Dense(hidden, activation='relu'))
    model.add(Dense(1, activation='linear'))
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss=MeanSquaredError(),
                  metrics=['accuracy'])
    model.summary()
    return model
```

- To reduce code duplication a function is defined for a sequential model with one hidden layer and an adjustable number of nodes
- 8 nodes are used in the hidden layer, the input layer has 6 nodes corresponding to the number of features in the full model, there is only one node in the output layer corresponding to the single target ROA

```python
# Extract model weights by layer
weights0, biases0 = model_f.layers[0].get_weights()
weights1, biases1 = model_f.layers[1].get_weights()
weights0, weights1
```

- The weights between each layer of the network
- The first array contains 6 arrays one for each input containing 8 weights one for each node in the hidden layer
- The second array contains the eight weight from the hidden layer to the output node

```python
# Define function for calculating influence of each input to output
def influence_f(weights0, inputs, h_nodes):
    influence = {}
    for i in range(inputs):
        influence[i] = np.sum(np.abs(weights0[i]))
    return influence
```

- The function above sums the absolute value of the weights for each input as a measure of feature importance
- The absolute value is used of each weight because some weights may be negative

```python
# Influence of inputs to output
# ['Loan Growth Rate', 'GDP Growth Rate', 'Total Assets', 'Interbank Rate', 'Liquidity', 'Credit Risk']
influence = influence_f(weights0, 6, 8)
print(f"""Loan Growth Rate: {influence[0]}
GDP Growth Rate: {influence[1]}
Total Assets: {influence[2]}
Interbank Rate: {influence[3]}
Liquidity: {influence[4]}
Credit Risk: {influence[5]}""")
```

## Time Series Split

- This section uses sci-kit learns TimeSeriesSplit function as an alternative technique for splitting the data into train and validation
    - It uses a rolling window starting at the beginning of the time series with a small train and validation set for the first split (test set in image below)
    - For the next split the train and test set move along the time series and the new training set includes the old test set and so on
    - The image below gives a visual representation of the process with 4 splits found at: https://github.com/koaning/scikit-lego/issues/70





```python
# Define time series split cross-validation function
def time_series_split(model, x, y, k, **kwargs):
    histories = []
    scores = []

    for train, test in TimeSeriesSplit(n_splits = k).split(x):
        histories.append(model.fit(x[train], y[train], validation_data = (x[test], y[test]),
                              **kwargs).history)
        scores.append(model.evaluate(x[test], y[test], verbose=0))
    return histories, scores
```

- The function above fits a model with Time Series Split using a specific number of splits 
- It collects the histories and scores from each split 

```python
# Define function for averaging losses across folds (5)
def mean_loss(k, history, loss, val_loss):
    mean_loss = []
    mean_val_loss = []
    for i in range(len(history[0][loss])):
        mean_loss.append((history[0][loss][i] + history[1][loss][i] + history[2][loss][i] + history[3][loss][i] + history[4][loss][i])/k)
        mean_val_loss.append((history[0][val_loss][i] + history[1][val_loss][i] + history[2][val_loss][i] + history[3][val_loss][i] + history[4][val_loss][i])/k)
    return mean_loss, mean_val_loss
```

- The function above calculates the average training and validation loss across each split, in this case 5 splits

```python
# Define function for averaging accuracy across folds (5)
def mean_accuracy(k, history, accuracy, val_accuracy):
    mean_accuracy = []
    mean_val_accuracy = []
    for i in range(len(history[0][accuracy])):
        mean_accuracy.append((history[0][accuracy][i] + history[1][accuracy][i] + history[2][accuracy][i] + history[3][accuracy][i] + history[4][accuracy][i])/k)
        mean_val_accuracy.append((history[0][val_accuracy][i] + history[1][val_accuracy][i] + history[2][val_accuracy][i] + history[3][val_accuracy][i] + history[4][val_accuracy][i])/k)
    return mean_accuracy, mean_val_accuracy
```

- The function above is identical to the previous function but calculates the average accuracy across each split

# Bayesian Linear Regression

- This section aims to analyze the research question using Bayesian linear regression and is structured as follows:
    - First the data is standardized to have mean 0 and standard deviation as 1 to help the algorithm.
    - Second full Bayesian model with all variables is specified. Tight priors are provided.
    - Third the model fit is investigated to check if the algorithm has converged. Results of the posterior distribution of each predictor are discussed.
    - Fourth alternative model without Total Assets is created and the two models are compared using Widely Applicable Information Criterion (WAIC).
    - Finally, we generate random data and investigate if our chosen model retrieves the true parameters, for sensitivity analysis.

## Standardized data

- Similarly as with neural networks, we help the algorithm by standardizing our variables with mean 0 and standard deviation 1
- For this we create function that standardized a variable

```python
def standardize(x):
    return (x-x.mean())/x.std()
ROA = standardize(df['ROA'])
LoanGtowth = standardize(df['Loan Growth Rate'])
GDP = standardize(df['GDP Growth Rate'])
TA = standardize(df['Total Assets'])
InterBank = standardize(df['Interbank Rate'])
Liquidity = standardize(df['Liquidity'])
CR = standardize(df['Credit Risk'])
```

```python
# store standartized variables in a Data Frame
df_stand = pd.DataFrame(list(zip(ROA, LoanGtowth, GDP, TA, InterBank, Liquidity, CR)),
               columns =['ROA', 'LG', 'GDP', 'TA', 'Interbank', 'Liq', 'CR'])
df_stand.head()
```

- We can see that indeed mean is 0 and std is 1

```python
df_stand.describe()
```

## Full Bayesian Model

```python
# specify the Baysian model
with pm.Model() as profitability_model:


      ## priors on slope coefficients
      constant = pm.Normal('constant', mu = 0.0, sd = 1.0)
      σ_prior = 0.1 #seatbelt prior 
      slope_LoanGrowth = pm.Normal('slope_LoanGrowth', 0, σ_prior)
      slope_GDP = pm.Normal('slope_GDP', 0, σ_prior)
      slope_TotalAssets = pm.Normal('slope_TotalAssets', 0, σ_prior)
      slope_InterBank = pm.Normal('slope_InterBank', 0, σ_prior)
      slope_Liquidity = pm.Normal('slope_Liquidity', 0, σ_prior)
      slope_CreditRisk = pm.Normal('slope_CreditRisk', 0, σ_prior)
      
      ## model
      μ = constant + slope_LoanGrowth * LoanGtowth + slope_GDP * GDP + slope_TotalAssets * TA +  slope_InterBank * InterBank + slope_Liquidity * Liquidity + slope_CreditRisk * CR
      σ = pm.HalfNormal('σ', 1)

      profitability = pm.Normal('profitability', μ, σ, observed=ROA) # we assume it is normally distributed as it is standartized
      # sample from the model:
      trace_normal = pm.sample(10000, tune=1000)
```

- Selection of priors:
  - Sigma cannot be negative and Half Normal distribution deals with that.

  - We choose conservative priors with zero expectation and relatively small standard deviations (tight around zero) to avoid potential over-fitting. More evidence in the data is needed before the slope coefficients are moved from zero.
Slope coefficients are allowed to be both positive and negatives, so Normal distribution is specified. Since the variables are standardized they are normally distributed.


```python
# Summarized model
pm.model_to_graphviz(profitability_model)
```

```python
with profitability_model:
  # to compare model prediction and draw 95% credible interval
  ppc_model = pm.sample_posterior_predictive(trace_normal, var_names=['profitability'])
  # store results
  posterior_profitability = az.from_pymc3(
    trace_normal,
    posterior_predictive = ppc_model)
```

- Checking the model fit

```python
# plot trace 
with profitability_model:
    pm.plot_trace(trace_normal)
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

- We see posterior distribution of the model estimated in the charts. Estimated chains provide rather similar information on the left, which is a good indicator. 

- Trace plots on the right also show positive signs of the model's behaviour. Figures are stationary for all variables, as there are no trend lines and lines are centered around mean value. Moreover, we observe condensed zig-zagging, as in small intervals the lines cover a wide range of posterior distribution. Indeed, we trust sampling process better if we draw samples in limited intervals that include a wider distribution. Finally, we want different chains to cover the same region and this is also observed in the plots. Hence, as all three conditions are satisfied we are rather confident that the algorithm used to draw samples from the posterior (MCMC) have worked. 

- We also plot the posterior distributions as histograms with 95% credibility interval:

```python
pm.plot_posterior(trace_normal, hdi_prob = 0.95);
```

- For demonstrating results, we can also show the prior and posterior distribution of a parameter. Since priors are the same for all variables, we show the two plots only for one variable, such as Liquidity

## Generate Data (Sensitivity Analysis)

- Next we generate data based on Model 1 parameters and check if it recovers these true parameters of the relationships. As we generate data ourselves, we know that are the true parameters.

```python
# mean slopes taken from the model 1 summary table
def generating_data(a1 = 0.138, a2 = 0.036, a3 = -0.202, a4 = -0.023, a5 = 0.307, a6 = -0.552, b = 0.0, n=len(df_stand)): 
    LoanG = pm.Normal.dist(0,1).random(size=n)
    GDP = pm.Normal.dist(0,1).random(size=n)
    TA = pm.Normal.dist(0,1).random(size=n)
    IntBank = pm.Normal.dist(0,1).random(size=n)
    Liquid = pm.Normal.dist(0,1).random(size=n)
    CredRis = pm.Normal.dist(0,1).random(size=n)
    noise = pm.Normal.dist(0,1).random(size=n)
    RoA = b + a1 * LoanG + a2 * GDP + a3 * TA + a4 * IntBank + a5 * Liquid + a6 * CredRis + noise 
    df_gen = pd.DataFrame({'RoA':RoA, 'LoanG':LoanG, 'GDP':GDP, 'TA':TA, 'IntBank':IntBank, 'Liquid':Liquid, 'CredRis':CredRis})
    return df_gen

df_generated = generating_data()
```

```python
# specify the Baysian model
with pm.Model() as profitability_model_generated:


      ## priors on slope coefficients
      constant = pm.Normal('constant', mu = 0.0, sd = 1.0)
      σ_prior = 0.1 #seatbelt prior 
      slope_LoanGrowth = pm.Normal('slope_LoanGrowth', 0, σ_prior)
      slope_GDP = pm.Normal('slope_GDP', 0, σ_prior)
      slope_TotalAssets = pm.Normal('slope_TotalAssets', 0, σ_prior)
      slope_InterBank = pm.Normal('slope_InterBank', 0, σ_prior)
      slope_Liquidity = pm.Normal('slope_Liquidity', 0, σ_prior)
      slope_CreditRisk = pm.Normal('slope_CreditRisk', 0, σ_prior)
      
      ## model
      μ = constant + slope_LoanGrowth * df_generated.LoanG + slope_GDP * df_generated.GDP + slope_TotalAssets * df_generated.TA +  slope_InterBank * df_generated.IntBank + slope_Liquidity * df_generated.Liquid + slope_CreditRisk * df_generated.CredRis
      σ = pm.HalfNormal('σ', 1)

      profitability = pm.Normal('profitability', μ, σ, observed=df_generated.RoA)
      # sample from the model:
      trace_normal_generated = pm.sample(10000, tune=1000)
```

```python
# Summarized model
pm.model_to_graphviz(profitability_model_generated)
```

```python
with profitability_model_generated:
  # to compare model prediction and draw 95% credible interval
  ppc_model_generated = pm.sample_posterior_predictive(trace_normal_generated, var_names=['profitability'])
  # store results
  posterior_profitability_generated = az.from_pymc3(
    trace_normal_generated,
    posterior_predictive = ppc_model_generated)
```

## Discussion of Results

The full neural network models display some variation between the train test split and time series split methods for training the models. The four most important features with the train test split are liquidity, GDP Growth Rate, Loan Growth Rate, and Credit Risk in that order. However, the order changes to Loan Growth Rate, Liquidity, Credit Risk, and GDP Growth Rate when using the time series split. For both methods the least two important features are Total Assets and the Interbank Rate in that order. 

The incomplete neural network models without total assets rank the remaining features in the same order for both splitting methods. The resulting order from most to least important is Credit Risk, Loan Growth Rate, GDP Growth Rate, Interbank Rate, and Liquidity. 

Interestingly, Liquidity is ranked as least important by the incomplete models in comparison to the full models which rank it as either most important or second most important. Liquidity in the incomplete models is lower than the Interbank Rate which is ranked lowest by the full models. Credit Risk is ranked fourth and third by the train test split and time series split full models, respectively. In comparison the incomplete models rank Credit Risk as most important followed by Loan Growth Rate which is ranked as third and first in importance by the full models. 

The random forest models with and without total assets rank the features in the same order besides not having total assets for the incomplete model. The resulting feature importance ranking is Credit Risk, Liquidity, Total Assets, Interbank Rate, GDP Growth Rate, and Loan Growth Rate. Liquidity was highly ranked by the full neural networks while Credit Risk was ranked slightly lower. This order of importance is switched by the random forest models. The Total Assets and the Interbank Rate are ranked higher by the random forest models than the neural networks. Loan Growth Rate is ranked as least important by the random forest models while it's ranked as most, second most, or third most important by the neural networks. 

When it comes the Bayesian regression, there is also some variation in terms of determinants importance between the full and incomplete model. The absolute mean effect of the input variable on the outcome variable is chosen in order to rank the determinants. For the full model Credit Risk, Liquidity, Total Assets and Loan Growth Rate (in that order) have the highest impact on ROA. GDP Growth Rate and Interbank Rate are approximated by smallest relationship with the outcome variable. The incomplete model shows that Credit Risk, Liquidity and Loan Growth Rate are of the highest importance, while GDP Growth Rate and Interbank Rate have again the weakest correlation effects. Overall, both models show that Credit Risk followed by Liquidity are ranked highest by the largest impact on profitability. 

The quarterly frequency of the time series data limits the dataset to only 147 observations for each variable. This is a relatively smaller dataset for training neural networks. Having access to monthly data for each variable may have improved the performance of the models, however, only securities and loans in bank credit are available in monthly frequency from the FRED. 

The addition of features may have improved the prediction power of the models. Bank capital was not available from the FRED to create a variable for the ratio of capital to total assets. A proxy for efficiency as the ratio between operational expenses gross revenues may have improved the model if data on these two variables for banks was available from the FRED.
