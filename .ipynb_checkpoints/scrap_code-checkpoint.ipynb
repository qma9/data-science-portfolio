```python
class DataFrames():

    def __init__(self, ticker, year_start, month_start, day_start, indicators):

        self.ticker = ticker
        self.year_start = year_start
        self.month_start = month_start
        self.day_start = day_start
        self.import_()
        if indicators:
            self.daily_return()
            self.sd()
            self.moving_avg()
            self.exp_moving_avg()
            self.on_balance_volume()
            self.bollinger_bands()
            self.relative_strength_index()

    def import_(self):
        self.df = pdr.get_data_yahoo(
            self.ticker,
            dt.datetime(self.year_start, self.month_start, self.day_start),
            dt.datetime.now())
        self.df.reset_index(inplace=True)

    def daily_return(self):
        self.df['Return'] = self.df['Close'] - self.df['Open']

    def sd(self):
        self.df['SD'] = self.df['Return'].rolling(window=20,
                                                  min_periods=0).std()

    def moving_avg(self):
        self.df['MA20'] = self.df['Close'].rolling(window=20,
                                                   min_periods=0).mean()

    def exp_moving_avg(self):
        self.df['EMA20'] = self.df['Close'].ewm(span=20, adjust=False).mean()

    def on_balance_volume(self):
        self.df['OBV'] = np.where(
            self.df['Close'] > self.df['Close'].shift(1), self.df['Volume'],
            np.where(self.df['Close'] < self.df['Close'].shift(1),
                     -self.df['Volume'], 0)).cumsum()

    def bollinger_bands(self):
        self.df['BB_upper'] = self.df['MA20'] + self.df['SD'] * 2
        self.df['BB_lower'] = self.df['MA20'] - self.df['SD'] * 2
        
    def relative_strength_index(self):
        self.df['RSI'] = pta.rsi(self.df['Close'], length = 14)
```

```python
import datetime
import pandas as pd
import numpy as np
import pandas_datareader.data as web 


class Security:
    def __init__(self, ticker, data_platform, start_date, end_date):
        self.ticker = ticker
        self.data_platform = data_platform
        self.start_date = start_date
        self.end_date = end_date
        self.df = None

    def fetch_stock_data(self):
        self.df = web.DataReader(self.ticker, self.data_platform, self.start_date, self.end_date)
        return self.df
```

```python
class Evaluation:

    def __init__(self, security):
        self.security = security

    def simple_moving_average(self, period):
        df = self.security
        return df['Adj Close'].rolling(period).mean()

    def exp_moving_average(self, period):
        df = self.security
        return df['Adj Close'].ewm(span=period).mean()

    def rsi(self, period):
        df = self.security
        delta = df['Adj Close'].diff()
        up = delta * 0
        down = up.copy()
        up[delta > 0] = delta[delta > 0]
        down[delta < 0] = -delta[delta < 0]
        up[up.index[period - 1]] = np.mean(up[:period])
        up = up.drop(up.index[:(period - 1)])
        down[down.index[period - 1]] = np.mean(down[:period])
        down = down.drop(down.index[:(period - 1)])
        rs = up.ewm(span=period - 1).mean() / down.ewm(span=period - 1).mean()
        return 100 - 100 / (1 + rs)
```

```python
def train_val_test(df, train, validation):

    features = np.array(
        df[['High', 'Low', 'Open', 'Return', 'SD', 'MA20', 'EMA20', 'RSI']])
    target = np.array(df['Close'])

    data_len = df.shape[0]
    print(f'Length of data = {data_len}')

    train_split = int(data_len * train)
    print(f'Lenght of training set = {train_split}')

    val_split = train_split + int(data_len * validation)
    print(f'Length of validation set = {int(data_len * validation)}')

    print(
        f'Length of test set = {data_len - train_split - int(data_len * validation)}'
    )

    x_train, x_val, x_test = features[:train_split], features[
        train_split:val_split], features[val_split:]
    y_train, y_val, y_test = target[:train_split], target[
        train_split:val_split], target[val_split:]

    return x_train, x_val, x_test, y_train, y_val, y_test
```
