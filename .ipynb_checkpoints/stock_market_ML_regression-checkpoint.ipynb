```python
from typing import Callable
import functools

ComposableFunction = Callable[[float], float]


class DataSet:
    """
    Import data through pandas datareader.

    Paramaters
    ----------
    ticker : str
        Stock ticker symbol
    year_start : int
        Starting year of the time series
    month_start : int
        Starting month of the time series
    day_start : int
        Starting day of the time series
    """

    def __init__(self, ticker, year_start, month_start, day_start):

        self.ticker = ticker
        self.year_start = year_start
        self.month_start = month_start
        self.day_start = day_start
        self.df = None

    def get_data(self):
        self.df = pdr.get_data_yahoo(
            self.ticker,
            dt.datetime(self.year_start, self.month_start, self.day_start),
            dt.datetime.now())
        self.df.reset_index(inplace=True)
        return self.df


class Indicators:
    """
    Calculates indicators and stores them in new columns.
    """

    def __init__(self, df):
        self.df = df

    def _compose(self, *functions: ComposableFunction) -> ComposableFunction:
        return functools.reduce(lambda f, g: lambda x: g(f(x)), functions)

    def calculate(self):
        composed_function = self._compose(
            self._daily_return, self._standard_deviation, self._moving_average,
            self._exponential_moving_average, self._on_balance_volume,
            self._bollinger_bands, self._relative_strength_index,
            self._average_true_range, self._average_directional_movement_index,
            self._commodity_channel_index, self._william_r_oscillator)
        self.df = composed_function(self.df)
        return self.df

    def _daily_return(self, df):
        df['Return'] = df['Close'] - df['Open']
        return self.df

    def _standard_deviation(self, df):
        df['SD'] = df['Return'].rolling(window=20, min_periods=0).std()
        return self.df

    def _moving_average(self, df):
        df['MA20'] = df['Close'].rolling(window=20, min_periods=0).mean()
        return self.df

    def _exponential_moving_average(self, df):
        df['EMA20'] = df['Close'].ewm(span=20, adjust=False).mean()
        return self.df

    def _on_balance_volume(self, df):
        df['OBV'] = np.where(
            df['Close'] > df['Close'].shift(1), df['Volume'],
            np.where(df['Close'] < df['Close'].shift(1), -df['Volume'],
                     0)).cumsum()
        return self.df

    def _bollinger_bands(self, df):
        df['BB_upper'] = df['MA20'] + df['SD'] * 2
        df['BB_lower'] = df['MA20'] - df['SD'] * 2
        return self.df

    def _average_true_range(self, df):
        true_range = pd.DataFrame(
            {
                'TR':
                np.fmax(np.abs(df['High'] - df['Low']),
                        np.abs(df['High'] - df['Close'].shift()),
                        np.abs(df['Low'] - df['Close'].shift()))
            },
            index=np.arange(len(df)))
        df['ATR'] = true_range.ewm(span=20, adjust=False).mean()
        return self.df

    def _average_directional_movement_index(self, df):
        dm_pos = pd.DataFrame(
            {
                'DM': np.fmax(df['Close'] - df['Close'].shift(),
                              np.zeros(len(df)))
            },
            index=np.arange(len(df)))
        dm_neg = pd.DataFrame(
            {
                'DM': np.fmin(df['Close'] - df['Close'].shift(),
                              np.zeros(len(df)))
            },
            index=np.arange(len(df)))
        di_pos = 100 * dm_pos['DM'].ewm(span=20,
                                        adjust=False).mean() / df['ATR']
        di_neg = 100 * dm_neg['DM'].ewm(span=20,
                                        adjust=False).mean() / df['ATR']
        df['ADMI'] = 100 * (di_pos - di_neg) / (di_pos + di_neg)
        return self.df

    def _commodity_channel_index(self, df):
        n = 20
        M = pd.DataFrame({'M': df['High'] + df['Low'] + df['Close']},
                         index=np.arange(len(df)))
        df['CCI'] = (M['M'] - M['M'].rolling(
            window=n, min_periods=0).mean()) / (0.015 * (np.sum(
                np.abs(M['M'] -
                       M['M'].rolling(window=n, min_periods=0).mean()))) / n)
        return self.df

    def _relative_strength_index(self, df):
        #df['RSI'] = pta.rsi(df['Close'], length=14)
        dm_pos = pd.DataFrame(
            {
                'DM': np.fmax(df['Close'] - df['Close'].shift(),
                              np.zeros(len(df)))
            },
            index=np.arange(len(df)))
        dm_neg = pd.DataFrame(
            {
                'DM': np.fmin(df['Close'] - df['Close'].shift(),
                              np.zeros(len(df)))
            },
            index=np.arange(len(df)))
        df['RSI'] = 100 - 100 / (
            1 + dm_pos['DM'].ewm(span=20, adjust=False).mean() /
            dm_neg['DM'].ewm(span=20, adjust=False).mean())
        return self.df

    def _william_r_oscillator(self, df):
        n = 20
        high_h = df['High'].rolling(n, min_periods=0).max()
        low_l = df['Low'].rolling(n, min_periods=0).min()
        df['W%R'] = -100 * ((high_h - df['Close']) / (high_h - low_l))
        return self.df
```

```python
# main
apple = DataSet('AAPL', 2006, 1, 1)
df_apple = Indicators(apple.get_data()).calculate()
df_apple
```

# Exploratory Data Analysis

```python
class Viz():
    """
    Generates visualizations of stock data.
    """
    def __init__(self):

        pass

    def figure(self, df, subplot_titles, title):

        fig = make_subplots(rows=2,
                                 cols=1,
                                 shared_xaxes=True,
                                 vertical_spacing=0.10,
                                 subplot_titles=subplot_titles,
                                 row_width=[0.2, 0.7])
        fig.add_trace(go.Candlestick(x=df['Date'],
                                          open=df["Open"],
                                          high=df["High"],
                                          low=df["Low"],
                                          close=df["Close"],
                                          name="OHLC"),
                           row=1,
                           col=1)
        fig.add_trace(go.Scatter(x=df['Date'],
                                      y=df["MA20"],
                                      marker_color='yellow',
                                      name="MA20"),
                           row=1,
                           col=1)
        fig.add_trace(go.Scatter(x=df['Date'],
                                      y=df["EMA20"],
                                      marker_color='lightblue',
                                      name="EMA20"),
                           row=1,
                           col=1)
        fig.add_trace(go.Bar(x=df['Date'],
                                  y=df['Volume'],
                                  marker_color='red',
                                  showlegend=False),
                           row=2,
                           col=1)
        fig.update_layout(title=title,
                               xaxis_tickfont_size=12,
                               yaxis=dict(
                                   title='Price ($/share)',
                                   titlefont_size=14,
                                   tickfont_size=12,
                               ),
                               autosize=False,
                               width=800,
                               height=500,
                               margin=dict(l=50, r=50, b=100, t=100, pad=4),
                               paper_bgcolor='LightSteelBlue',
                         plot_bgcolor='LightSteelBlue')
        fig.update(layout_xaxis_rangeslider_visible=True)
        return fig
```

# Feature Engineering

```python
class Normalize:
    """
    
    """

    def __init__(self, df):
        self.df = df

    def scale(self):
        features = np.array(self.df.loc[:, (self.df.columns != 'Date') &
                                        (self.df.columns != 'Close')])
        target = np.array(self.df['Close'])
        scaler_x = MinMaxScaler()
        scaler_y = MinMaxScaler()
        features_scaled = scaler_x.fit_transform(features)
        target_scaled = scaler_y.fit_transform(target.reshape(-1, 1))
        self.df = pd.DataFrame(
            np.concatenate((target_scaled, features_scaled), axis=1),
            index=self.df.loc[:, 'Date'],
            columns=[
                np.concatenate([
                    np.array(['Close'], str), df_apple.columns.values[
                        (df_apple.columns.values != 'Close')
                        & (df_apple.columns.values != 'Date')]
                ],
                               axis=0)
            ])
        return self.df, scaler_x, scaler_y
```

```python
# Define function for reshaping data into samples (rows), timesteps (look back), and features (columns)
# y is 2D

def reshape_2d(data, n_past, n_future):
    x, y = [], []
    
    for i in range(n_past, len(data) - n_future + 1):
        x.append(data[i - n_past:i, 0:data.shape[1]])
        y.append(data[i + n_future - 1:i + n_future, 0:data.shape[1]])
        
    x, y = np.array(x), np.array(y)
    return x, y
```

```python
class Reshape:
    """
    
    """

    def __init__(self, df):
        self.df = df

    def reshape(self, data, n_past, n_future):
        x = np.array([
            data[i - n_past:i, 0:data.shape[1]]
            for i in range(n_past,
                           len(data) - n_future + 1)
        ])
        y = np.array([data[i + n_future - 1:i + n_future, 0:data.shape[1]]]
                     for i in range(n_past,
                                    len(data) - n_future + 1))

        return x, y
```

```python
class DataSplit:
    """
    
    """

    def __init__(self, df):
        self.df = df

    def split(self, train, validation):
        features = np.array(self.df.loc[:, (self.df.columns != 'Date') &
                                        (self.df.columns != 'Close')])
        target = np.array(self.df['Close'])

        data_len = self.df.shape[0]
        print(f'Length of data = {data_len}')

        train_split = int(data_len * train)
        print(f'Lenght of training set = {train_split}')

        val_split = train_split + int(data_len * validation)
        print(f'Length of validation set = {int(data_len * validation)}')

        print(f'Length of test set = {data_len - train_split - int(data_len * validation)}')

        x_train, x_val, x_test = features[:train_split], features[
            train_split:val_split], features[val_split:]
        y_train, y_val, y_test = target[:train_split], target[
            train_split:val_split], target[val_split:]

        return x_train, x_val, x_test, y_train, y_val, y_test
```

```python
# main
x_train, x_val, x_test, y_train, y_val, y_test = DataSplit(
    df_apple).split(0.8, 0.11)
```

# Modelling

```python
class Modelling:

    def model_fit(self, input_layer, hidden_layer, learning_rate, x_train,
                  y_train, epochs, validation_data, x_test, scaler_y):
        model = Sequential()
        model.add(InputLayer(input_layer))
        model.add(Dense(hidden_layer, activation='relu'))
        model.add(Dense(1, activation='linear'))
        model.compile(optimizer=Adam(learning_rate=learning_rate),
                      loss=MeanSquaredError(),
                      metrics=['accuracy'])
        model.summary()
        history = model.fit(x_train,
                            y_train,
                            epochs=epochs,
                            validation_data=validation_data)
        predictions_scaled = model.predict(x_test)
        predictions = scaler_y.inverse_transform(predictions_scaled.reshape(-1, 1)).flatten()
        return model, history, predictions
```

```python
# main
model, history, predictions = Modelling().model_fit(input_layer=13,
                                                    hidden_layer=16,
                                                    learning_rate=0.01,
                                                    x_train=x_train,
                                                    y_train=y_train,
                                                    epochs=35,
                                                    validation_data=(x_val, y_val),
                                                    x_test=x_test,
                                                    scaler_y=scaler_y)
```

```python
class PlotHistoryPredictions:

    def __init__(self, df, history, predictions):
        self.df = df
        self.history = history
        self.predictions = predictions

    def plot(self):

        # Loss figure
        fig1, ax = plt.subplots(figsize=(8, 4))
        ax.plot(range(len(self.history.history['loss'])),
                self.history.history['loss'],
                label="Training Loss")
        ax.plot(range(len(self.history.history['val_loss'])),
                self.history.history['val_loss'],
                label="Validation Loss")
        ax.set_xlabel('Epochs')
        ax.set_ylabel('Loss')
        ax.legend(loc='best')

        # Accuracy figure
        fig2, ax = plt.subplots(figsize=(8, 4))
        ax.plot(range(len(self.history.history['accuracy'])),
                self.history.history['accuracy'],
                label="Training Accuracy")
        ax.plot(range(len(self.history.history['val_accuracy'])),
                self.history.history['val_accuracy'],
                label="Validation Accuracy")
        ax.set_xlabel('Epochs')
        ax.set_ylabel('Loss')
        ax.legend(loc='best')

        # Predicitons figure
        fig3, ax = plt.subplots(figsize=(9, 4))
        ax.plot(self.df.loc[len(self.df) - len(self.predictions):, 'Date'],
                self.predictions,
                c='r',
                label='Predictions')
        ax.plot(self.df.Date, self.df['Close'], c='b', label='Actuals')
        ax.set_xlabel('Date')
        ax.set_ylabel('Decimal')
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

        return fig1, fig2, fig3
```

```python
# main
fig1, fig2, fig3 = PlotHistoryPredictions(df_apple, history, predictions).plot()
```
